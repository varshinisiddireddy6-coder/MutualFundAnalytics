```python
import pandas as pd

perf_df = pd.read_csv("../data/raw/07_scheme_performance.csv")

perf_df.head()
```

```python
perf_df.shape
```

```python
perf_df.sort_values(
    by='sharpe_ratio',
    ascending=False
)[[
    'scheme_name',
    'fund_house',
    'sharpe_ratio'
]].head(10)
```

```python
import matplotlib.pyplot as plt

top10 = perf_df.sort_values(
    by='sharpe_ratio',
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

plt.bar(
    top10['scheme_name'],
    top10['sharpe_ratio']
)

plt.xticks(rotation=90)
plt.title("Top 10 Funds by Sharpe Ratio")
plt.ylabel("Sharpe Ratio")

plt.tight_layout()
plt.show()
```

```python
perf_df.sort_values(
    by='alpha',
    ascending=False
)[[
    'scheme_name',
    'alpha'
]].head(10)
```

```python
top_alpha = perf_df.sort_values(
    by='alpha',
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

plt.bar(
    top_alpha['scheme_name'],
    top_alpha['alpha']
)

plt.xticks(rotation=90)
plt.title("Top 10 Funds by Alpha")
plt.ylabel("Alpha")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    perf_df['std_dev_ann_pct'],
    perf_df['return_5yr_pct']
)

plt.xlabel("Risk (Std Dev %)")
plt.ylabel("5 Year Return %")
plt.title("Risk vs Return Analysis")

plt.grid(True)
plt.show()
```
